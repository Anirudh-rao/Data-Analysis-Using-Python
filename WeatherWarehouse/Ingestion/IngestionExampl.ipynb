```python
import pandas as pd
import requests
import os
from dotenv import load_dotenv
import io

load_dotenv()
```

```python
# Getting API keys
ApiKey = os.getenv('WeatherAPI')
```

```python
CitiesJson = pd.read_json('/Users/rookr/Documents/Data-Ingestion/WeatherWarehouse/Files/LocationIndia.json')
df = CitiesJson.head(10)
df
```

```python
# Api Ingestion


for index,row in df.iterrows():
    lat = row['lat']
    lon = row['lng']
    url = f'https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={ApiKey}/json'
    r = requests.get(url).json()
    # df = pd.json_normalize(r)

r


```
